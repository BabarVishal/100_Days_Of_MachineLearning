```python
df = df.drop_duplicates()
df
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
lists = ['salary', 'experience_years']
for list in lists:
    df[list] = df[list].fillna(df[list].median())
```

```python
df.isnull().sum()
```

```python
cats = ['city', 'education']
for cat in cats:
    df[cat] = df[cat].fillna(df[cat].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df = df.loc[:, ['age', 'salary', 'experience_years', 'score', 'review', 'working_hours']]
df.head()
```

```python
# Feature Scaling. StandardScaler -> KNN Algo. ..........................................................................................................
# min = 0 and STD = 1

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
          df.drop('review', axis=1),
          df['review'], 
          test_size = 0.3, 
          random_state = 1
)

x_train.shape , x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
data = StandardScaler()

data.fit(x_train)

x_train_scale = data.transform(x_train)
x_test_scale = data.transform(x_test)

x_train_scale = pd.DataFrame(x_train_scale, columns=x_train.columns)
x_test_scale = pd.DataFrame(x_test_scale, columns=x_test.columns)
```

```python
np.round(x_train.describe(), 1)
```

```python
np.round(x_train_scale.describe(), 1)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4))   # 1 row, 2 columns
# Before Scaling
sns.scatterplot(x=x_train['age'], y=x_train['salary'], ax=ax[0])
ax[0].set_title("Before Scaling")

# After Scaling
sns.scatterplot(x=x_train_scale['age'], y=x_train_scale['salary'], ax=ax[1])
ax[1].set_title("After Scaling")

plt.tight_layout()
plt.show()
```

```python
# Encoding Categorical Data.  Label Encoding -> for Output Col...........................................................................................
df.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    df.drop('review', axis=1),
    df['review'],
    test_size=0.3,
    random_state=1
)

x_train.shape, x_test.shape
```

```python
from sklearn.preprocessing import LabelEncoder
data = LabelEncoder()

data.fit(y_train)

y_train_scale = data.transform(y_train)
y_test_scale = data.transform(y_test)

y_train_scale = pd.DataFrame(y_train_scale, columns=["review"])
y_test_scale = pd.DataFrame(y_test_scale, columns=["review"])
```
