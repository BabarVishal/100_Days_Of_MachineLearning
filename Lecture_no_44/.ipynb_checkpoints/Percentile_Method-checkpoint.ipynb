```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('placement_normal_cgpa.csv')
df.head()
```

```python
sns.histplot(df['cgpa'], kde=True)
```

```python
sns.boxplot(df['cgpa'])
```

```python
upper_limit = df['cgpa'].quantile(0.99)
upper_limit
```

```python
lower_limit = df['cgpa'].quantile(0.01)
lower_limit
```

```python
df_new = df[(df['cgpa'] <= upper_limit) & (df['cgpa'] >= lower_limit)]
```

```python
df_new.head()
```

```python
sns.boxplot(df_new['cgpa'])
```

```python
df['cgpa'] = np.where(
    df['cgpa'] >= upper_limit,
    upper_limit, 
    np.where(
        df['cgpa'] <= lower_limit, 
        lower_limit, 
        df['cgpa']
    )
)
```

```python
df.shape
```

```python
sns.boxplot(df['cgpa'])
```
