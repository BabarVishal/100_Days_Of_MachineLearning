```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'} 
web_page = requests.get('https://www.cricbuzz.com/cricket-schedule/upcoming-series/all', headers=headers).text
```

```python
soup = BeautifulSoup(web_page, 'lxml')
print(soup.prettify())
```

```python
soup.find_all('div', class_='py-5 flex justify-between')[0].text.strip()
```

```python
data = soup.find_all('div', class_='py-5 flex justify-between')

list = []
for i in data: 
    list.append(i.text.strip())
```

```python
list
```

```python
df = pd.DataFrame(list)
df.rename(columns={0: 'data'})
```

```python
df.to_csv('data')
```
