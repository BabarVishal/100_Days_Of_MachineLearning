```python
#Convert DataType...........................................................................................................
data['id'] = data['id'].astype('Int64')
data['age'] = data['age'].astype('Int64')
data['joining_year'] = data['joining_year'].astype('Int64')
data['experience_years'] = data['experience_years'].astype('Int64')
```

```python
data.info()
```

```python
# Handling missing values......................................................................................................
data.isnull().sum()
```

```python
data['id'] = data['id'].fillna(-1)
# data = data.dropna(subset=['id'])
```

```python
data.isnull().sum()
```

```python
col_data = ['age', 'salary', 'experience_years', 'score', 'joining_year']
for col in col_data:
    data[col] = data[col].fillna(data[col].median())
```

```python
data.isnull().sum()
```

```python
cat_col = ['gender', 'city', 'department']
for col in cat_col: 
    data[col] = data[col].fillna(data[col].mode()[0])
```
