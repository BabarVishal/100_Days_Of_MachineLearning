```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('education_dataset_100.csv')
df.head()
```

```python
df = df.iloc[ : , 2: ]
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    df.iloc[ : , 0:2], 
    df.iloc[ : , -1 ],
    test_size=0.2,
    random_state=1
)

x_train.shape , x_test.shape
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Make sure THIS x_train contains original text values (strings)
obj = OrdinalEncoder(categories=[
    ['Poor', 'Average', 'Good'],
    ['HS', 'UG', 'PG']
])

obj.fit(x_train)

x_train = obj.transform(x_train)
x_test = obj.transform(x_test)

x_train = pd.DataFrame(x_train, columns=['review', 'education'])
x_test = pd.DataFrame(x_test, columns=['review', 'education'])

x_train
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# Fit on training data only
le.fit(y_train)

# Transform
y_train = le.transform(y_train)
y_test = le.transform(y_test)

# Convert to DataFrame
y_train = pd.DataFrame(y_train, columns=['purchased'])
y_test = pd.DataFrame(y_test, columns=['purchased'])

y_train
```
