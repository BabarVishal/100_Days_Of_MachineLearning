```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv('analysis_data_with_missing.csv')
df.head()
```

```python
df.isnull().mean()*100
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.compose import ColumnTransformer 
from sklearn.impute import SimpleImputer
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    df.drop('experience_years', axis=1), 
    df['experience_years'], 
    test_size=0.2, 
    random_state=3
)
x_train.shape, x_test.shape
```

```python
num_data = SimpleImputer(strategy='median')
cat_data = SimpleImputer(strategy='most_frequent')
```

```python
tfr = ColumnTransformer([
    ('num_data', num_data, ['age', 'salary']),
    ('cat_data', cat_data, ['gender', 'city'])
], remainder='passthrough')
```

```python
x_train = tfr.fit_transform(x_train)
x_test = tfr.transform(x_test)
```

```python
y_imp = SimpleImputer(strategy='median')
y_train = y_imp.fit_transform(y_train.to_numpy().reshape(-1,1))
y_test = y_imp.transform(y_test.to_numpy().reshape(-1,1))
```
