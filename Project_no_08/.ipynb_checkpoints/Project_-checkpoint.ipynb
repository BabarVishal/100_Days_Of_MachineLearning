```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv('analysis_data_with_missing.csv')
df.head()
```

```python
df.isnull().mean() * 100
```

```python
from sklearn.impute import SimpleImputer 
from sklearn.compose import ColumnTransformer 
```

```python
cat_values = SimpleImputer(strategy='most_frequent')
num_values = SimpleImputer(strategy='median')

Ctf = ColumnTransformer([
    ('cat_values', cat_values, ['gender', 'city']), 
    ('num_values', num_values, ['age', 'salary', 'experience_years'])
], remainder='passthrough')
```

```python
Ctf = Ctf.fit_transform(df)
Ctf
```

```python
df = pd.DataFrame(
    Ctf, 
    columns=['gender', 'city', 'age', 'salary', 'experience_years']
)
df
```

```python
df.isnull().mean()
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    df.drop('salary', axis=1), 
    df['salary'], 
    test_size=0.2, 
    random_state=2
)
x_train.shape, x_test.shape
```
