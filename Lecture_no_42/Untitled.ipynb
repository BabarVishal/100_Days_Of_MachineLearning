```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('placement_normal_cgpa.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(16,5))

plt.subplot(1,2,1)
sns.histplot(df['cgpa'], kde=True)

plt.subplot(1,2,2)
sns.histplot(df['placement_exam_marks'], kde=True)

plt.show()
```

```python
df['cgpa_Zscore'] = (df['cgpa'] - df['cgpa'].mean()) / (df['cgpa'].std()) #add new col
```

```python
df[(df['cgpa_Zscore'] > 3) | (df['cgpa_Zscore'] < -3)]  #check how many outliers
```
