```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import scipy.stats as stats # for QQ plot
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.model_selection import cross_val_score 
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import r2_score
from sklearn.preprocessing import PowerTransformer
```

```python
df = pd.read_csv('concrete_strength_500rows.csv')
df.head()
```

```python
df.describe()
```

```python
x = df.drop(columns=['Strength'])
y = df.iloc[ : , -1]

x_train, x_test, y_train, y_test = train_test_split(
    x, 
    y, 
    test_size=0.3, 
    random_state=1
)
```

```python
# Applying Regression without transformation 
lr = LinearRegression() 
lr.fit(x_train, y_train) 
y_pred = lr.predict(x_test)
r2_score(y_test, y_pred)
```

```python
# cross checking with cross val score 
le = LinearRegression() 
np.mean(cross_val_score(lr, x, y, scoring='r2'))
```

```python
# Plotting the distplots without anyt transformation 
for col in x_train.columns:
    plt.figure(figsize=(12, 5))

    # Distribution Plot
    plt.subplot(1, 2, 1)
    sns.histplot(x_train[col], kde=True)
    plt.title(f"Distribution of {col}")

    # Q-Q Plot
    plt.subplot(1, 2, 2)
    stats.probplot(x_train[col], dist="norm", plot=plt)
    plt.title(f"Q-Q Plot of {col}")

    plt.tight_layout()
    plt.show()
```

```python
# Power Transform
# Box-cos Transform...................................................................
pt = PowerTransformer(method='box-cox')

x_train_transformed = pt.fit_transform(x_train+0.000001)
x_test_transformed = pt.transform(x_test+0.000001)

pd.DataFrame({'cols': x_train.columns, 'box_col_lambdas': pt.lambdas_})
```

```python
lr = LinearRegression() 
lr.fit(x_train_transformed, y_train) 

y_red2 = lr.predict(x_test_transformed)
r2_score(y_test, y_red2)
```

```python
# Apply Yeo-johnson transform 
# Power Transformer
pt1 = PowerTransformer()

x_train_transformed2 = pt1.fit_transform(x_train)
x_test_transformed2 = pt1.transform(x_test)

# Linear Regression
lr = LinearRegression()
lr.fit(x_train_transformed2, y_train)

# Prediction
y_pred3 = lr.predict(x_test_transformed2)
print("R2 Score:", r2_score(y_test, y_pred3))

# Lambda values for each column
lambda_df = pd.DataFrame({
    'col': x_train.columns,
    'Yeo_Johnson_lambda': pt1.lambdas_
})

lambda_df
```
