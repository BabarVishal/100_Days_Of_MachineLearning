```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('student_placement_data (1).csv')
df.head()
```

```python
plt.figure(figsize=(14, 5))

plt.subplot(1,2,1)
sns.histplot(df['cgpa'], kde=True) 

plt.subplot(1,2,2) 
sns.histplot(df['placement_exam_marks'], kde=True)
```

```python
df['placement_exam_marks'].skew()
```

```python
df['placement_exam_marks'].describe()
```

```python
sns.boxplot(df['placement_exam_marks'])
```

```python
# Find the IQR 
perce_25 = df['placement_exam_marks'].quantile(0.25)
perce_75 = df['placement_exam_marks'].quantile(0.75)
```

```python
iqr = perce_75 - perce_25
iqr
```

```python
upper_limit = perce_75 + 1.5 * iqr 
lower_limit = perce_25 - 1.5 * iqr 
```

```python
# finding outliers 
df[df['placement_exam_marks'] > upper_limit]
```

```python
df[df['placement_exam_marks'] < lower_limit]
```

```python
# triming 
df_new = df[df['placement_exam_marks'] < upper_limit]
```

```python
df_new.shape
```

```python
plt.figure(figsize=(14, 8))

plt.subplot(2,2,1)
sns.histplot(df['cgpa'], kde=True) 

plt.subplot(2,2,2) 
sns.histplot(df['placement_exam_marks'], kde=True) 

plt.subplot(2,2,3)
sns.histplot(df_new['cgpa'], kde=True) 

plt.subplot(2,2,4) 
sns.histplot(df_new['placement_exam_marks'], kde=True) 

plt.show()
```

```python
# capping 
new_df_cap = df.copy() 
```

```python
new_df_cap['placement_exam_marks'] = np.where(
    new_df_cap['placement_exam_marks']  > upper_limit, 
    upper_limit, 
    np.where(
        new_df_cap['placement_exam_marks'] < lower_limit, 
        lower_limit, 
        new_df_cap['placement_exam_marks']
    )
)
```
