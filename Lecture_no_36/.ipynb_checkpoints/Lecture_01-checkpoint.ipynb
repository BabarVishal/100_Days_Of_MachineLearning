```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.impute import SimpleImputer 
from sklearn.compose import ColumnTransformer 
```

```python
df = pd.read_csv('random_numeric_missing_1000rows.csv')
df.head()
```

```python
df.isnull().mean()
```

```python
simpalImputer = SimpleImputer(strategy='median')
```

```python
trf = ColumnTransformer([
    ('simpalImputer', simpalImputer, ['col1', 'col2', 'col3', 'col4', 'col5'])
], remainder='passthrough')
```

```python
trf.fit(df)
```

```python
trf.named_transformers_['simpalImputer'].statistics_
```

```python
x = trf.transform(df)
```
