```python
data =['salary', 'experience_years', 'score']
for col in data:
      df[col] = df[col].fillna(df[col].median())
```

```python
df.isnull().sum()
```

```python
citys = ['city']
for city in citys:
    df[city] = df[city].fillna(df[city].mode()[0])
```
