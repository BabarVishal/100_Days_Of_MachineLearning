```python
import pandas as pd 
import numpy as np 
```

```python
# Handling Date Values..........................................................
date = pd.read_csv('first_file.csv')
date.head()
```

```python
date.info()
```

```python
date['Date'] = pd.to_datetime(date['Date'])
```

```python
date.info()
```

```python
date.head()
```

```python
date['date_year'] = date['Date'].dt.year
```

```python
date['date_month'] = date['Date'].dt.month
```

```python
date['date_monthName'] = date['Date'].dt.month_name()
```

```python
date['date_day'] = date['Date'].dt.day
```

```python
date['date_dayOfWeek'] = date['Date'].dt.dayofweek
```

```python
date['date_week_name'] = date['Date'].dt.day_name()
```

```python
date['date_is_weekend'] = np.where(date['date_week_name'].isin(['Sunday', 'Saturday']), 1, 0)
```

```python
date['quater'] = date['Date'].dt.quarter
```

```python
date.head()
```

```python
# Handling Time Variables....................................
time = pd.read_csv('second_file.csv')
time.head()
```

```python
time['DateTime'] = pd.to_datetime(time['DateTime'])
```

```python
time.info()
```

```python
time['hour'] = time['DateTime'].dt.hour
time['min'] = time['DateTime'].dt.minute
time['sec'] = time['DateTime'].dt.second
time['time'] = time['DateTime'].dt.time
```
