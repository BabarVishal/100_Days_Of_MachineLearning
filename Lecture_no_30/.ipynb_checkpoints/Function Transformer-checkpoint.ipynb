```python
import pandas as pd
import numpy as np

import scipy.stats as stats

import matplotlib.pyplot as plt
import seaborn as sns 

from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score 
from sklearn.model_selection import cross_val_score 

from sklearn.linear_model import LogisticRegression 
from sklearn.tree import DecisionTreeClassifier 

from sklearn.preprocessing import FunctionTransformer 
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv('titanic_custom_100rows.csv', usecols=['Age', 'Fare', 'Survived'])
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
x = df.iloc[ : , 1:3]
y = df.iloc[ : , 0]
```

```python
# train test split 
x_train, x_test, y_train, y_test = train_test_split(
    x, 
    y, 
    test_size=0.2, 
    random_state=1
)
```

```python
plt.figure(figsize=(14, 4))

plt.subplot(121)
sns.histplot(x_train['Age'], kde=True)
plt.title('Age PDF')

plt.subplot(122)
stats.probplot(x_train['Age'], dist='norm', plot=plt)
plt.title('Age QQ plot')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 4))

plt.subplot(121)
sns.histplot(x_train['Fare'], kde=True)
plt.title('Fare PDF')

plt.subplot(122)
stats.probplot(x_train['Fare'], dist='norm', plot=plt)
plt.title('Fare QQ Plot')

plt.tight_layout()
plt.show()
```

```python
clf = LogisticRegression()
clf2 = DecisionTreeClassifier() 

clf.fit(x_train, y_train)
clf2.fit(x_train, y_train)

y_pred = clf.predict(x_test)       # Logistic Regression predictions
y_pred1 = clf2.predict(x_test)     # Decision Tree predictions

print("LR Accuracy:", accuracy_score(y_test, y_pred))
print("DT Accuracy:", accuracy_score(y_test, y_pred1))
```

```python
# 1. Transformer
trf = FunctionTransformer(func=np.log1p)

# 2. Transform train and test properly
x_train_trans = trf.fit_transform(x_train)
x_test_trans = trf.transform(x_test)

# 3. Models
clf = LogisticRegression(max_iter=200)
clf2 = DecisionTreeClassifier()

# 4. Fit models
clf.fit(x_train_trans, y_train)
clf2.fit(x_train_trans, y_train)

# 5. Predict on TEST data (correct way)
y_pred = clf.predict(x_test_trans)
y_pred1 = clf2.predict(x_test_trans)

# 6. Accuracy
print("LR Accuracy:", accuracy_score(y_test, y_pred))
print("DT Accuracy:", accuracy_score(y_test, y_pred1))
```

```python
plt.figure(figsize=(14, 4))

plt.subplot(121)
stats.probplot(x_train['Fare'], dist='norm', plot=plt)
plt.title('Before Fare QQ Plot')

plt.subplot(122)
stats.probplot(x_train_trans['Fare'], dist='norm', plot=plt)
plt.title('After Fare QQ Plot')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 4))

plt.subplot(121)
stats.probplot(x_train['Age'], dist='norm', plot=plt)
plt.title('Before Fare QQ Plot')

plt.subplot(122)
stats.probplot(x_train_trans['Age'], dist='norm', plot=plt)
plt.title('After Fare QQ Plot')

plt.tight_layout()
plt.show()
```

```python
def apply_transform(transform):
    
    # Separate features and target
    X = df.iloc[:, 1:3]    # columns 1 and 2 (example: Age, Fare)
    y = df.iloc[:, 0]      # column 0 (example: Survived)

    # Create a ColumnTransformer: apply transform ONLY on 'Fare'
    trf = ColumnTransformer(
        transformers=[
            ('log', FunctionTransformer(transform), ['Fare'])
        ],
        remainder='passthrough'
    )

    # Fit-transform X
    X_trans = trf.fit_transform(X)

    # Model
    clf = LogisticRegression(max_iter=300)

    # Print accuracy using cross-validation
    print("Accuracy:", np.mean(
        cross_val_score(clf, X_trans, y, scoring='accuracy', cv=10)
    ))

    # Plot Before vs After Transform
    plt.figure(figsize=(14, 4))

    # Before transform
    plt.subplot(121)
    stats.probplot(X['Fare'], dist='norm', plot=plt)
    plt.title('Fare Before Transform')

    # After transform
    plt.subplot(122)
    stats.probplot(X_trans[:, 0], dist='norm', plot=plt)
    plt.title('Fare After Transform')

    plt.tight_layout()
    plt.show()
```

```python
apply_transform(lambda x : x)
```

```python
apply_transform(lambda x : x ** 2)
```

```python
apply_transform(lambda x : x ** 3)
```

```python
apply_transform(lambda x : x **1/2)
```
